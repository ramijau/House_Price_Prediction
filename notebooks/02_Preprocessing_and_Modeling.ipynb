# Preprocessing and Modeling

```python
import pandas as pd
from pandas import Series
import pandas.core.algorithms as algos
import numpy as np
import seaborn as sns
import os
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats.stats as stats
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import traceback
import re
import string
import warnings
import collections
import shap
from sklearn import __version__ as sklearn_version
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV, learning_curve
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score, classification_report, f1_score, confusion_matrix, precision_recall_curve
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, log_loss, auc, roc_curve, roc_auc_score
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, f_regression, f_classif
import datetime
import time

from library.sb_utils import save_file
```

```python
# import data
hp = pd.read_csv('../data/house_prices.csv')
hp.head()
```

```python
hp.info()
```

### Train/Test Split

```python
# one hot encoding
object_types = list(hp.select_dtypes('object').columns)
hp_encoded = pd.get_dummies(hp, columns=object_types)
hp_encoded.shape
```

```python
X = hp_encoded.drop(columns='SalePrice')
y = hp_encoded['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=47)
```

```python
X_train.info()
```

#### Impute missing values and scale data

```python
X_defaults_median = X_train.median()
X_defaults_median
```

```python
X_tr = X_train.fillna(X_defaults_median)
X_te = X_test.fillna(X_defaults_median)
```

```python
scaler = StandardScaler()
scaler.fit(X_tr)
scaler.get_feature_names_out(X_tr.columns)
X_tr_scaled = scaler.transform(X_tr)
X_te_scaled = scaler.transform(X_te)
```

```python
X_tr_scaled = pd.DataFrame(X_tr_scaled, columns = X_tr.columns)
X_te_scaled = pd.DataFrame(X_te_scaled, columns = X_tr.columns)
```

### Random Forest Regressor

```python
# random forest regressor model
rfr = RandomForestRegressor(random_state=47)
```

```python
rfr.fit(X_tr_scaled, y_train)
```

```python
y_tr_pred = rfr.predict(X_tr_scaled)
y_te_pred = rfr.predict(X_te_scaled)
```

```python
# r2 score
print('Train r2 score: {}'.format(r2_score(y_train, y_tr_pred)))
print('Test r2 score: {}'.format(r2_score(y_test, y_te_pred)))
```

```python
# mean absolute error
print('Train mean absolute error: {}'.format(mean_absolute_error(y_train, y_tr_pred)))
print('Test mean absolute error: {}'.format(mean_absolute_error(y_test, y_te_pred)))
```

#### Grid Search Cross Validation

```python
# grid parameters
n_est = [int(n) for n in np.logspace(start=1, stop=3, num=20)]
grid_params = {'n_estimators': n_est,
               'max_depth': range (5,20,5),}
grid_params
```

```python
rf_grid_cv = GridSearchCV(rfr, 
                          param_grid=grid_params, 
                          cv=5, 
                          n_jobs=-1,
                          verbose=True)
```

```python
start_time = time.time()
rf_grid_cv.fit(X_tr_scaled, y_train)
#print("--- %s seconds ---" % (time.time() - start_time))
print('Runtime: {} seconds'.format(round(time.time() - start_time, 3)))
```

```python
print('Best Score: {}'.format(rf_grid_cv.best_score_))
print('Best Parameters: {}'.format(rf_grid_cv.best_params_))
```

```python
rf_best = rf_grid_cv.best_estimator_
```

```python
y_tr_pred = rf_best.predict(X_tr_scaled)
y_te_pred = rf_best.predict(X_te_scaled)
# r2 score
print('Train r2 score: {}'.format(r2_score(y_train, y_tr_pred)))
print('Test r2 score: {}'.format(r2_score(y_test, y_te_pred)))
```

```python
rf_best_cv_results = cross_validate(rf_best, X_tr_scaled, y_train, cv=5)
rf_best_scores = rf_best_cv_results['test_score']
rf_best_scores
```

```python
np.mean(rf_best_scores), np.std(rf_best_scores)
```

```python
plt.subplots(figsize=(5, 3))
imps = rf_best.feature_importances_
rf_feat_imps = pd.Series(imps, index=X_train.columns).sort_values(ascending=False)
rf_feat_imps[:10].plot(kind='bar')
plt.xlabel('features')
plt.ylabel('importance')
plt.title('Best random forest regressor feature importances');
```

```python
rf_neg_mae = cross_validate(rf_best, X_tr_scaled, y_train, 
                            scoring='neg_mean_absolute_error', cv=5, n_jobs=-1)
rf_neg_mae
```

```python
rf_mae_mean = np.mean(-1 * rf_neg_mae['test_score'])
rf_mae_std = np.std(-1 * rf_neg_mae['test_score'])
rf_mae_mean, rf_mae_std
```

```python
mean_absolute_error(y_test, rf_best.predict(X_te_scaled))
```

```python
fractions = [.2, .25, .3, .35, .4, .45, .5, .6, .75, .8, 1.0]
train_size, train_scores, test_scores = learning_curve(rf_best, 
                                                       X_tr_scaled, y_train, 
                                                       train_sizes=fractions)
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)
test_scores_std = np.std(test_scores, axis=1)
```

```python
plt.subplots(figsize=(10, 5))
plt.errorbar(train_size, test_scores_mean, yerr=test_scores_std)
plt.xlabel('Training set size')
plt.ylabel('CV scores')
plt.title('Cross-validation score as training set size increases')
plt.savefig('../images/CV_increasing_training_set.png', bbox_inches='tight')
plt.show()
```
